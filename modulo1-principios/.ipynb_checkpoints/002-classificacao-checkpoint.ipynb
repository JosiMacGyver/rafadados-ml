```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Problema:
 - Os clientes estão fechando a conta em um banco
 - precisamos usar os dados  para saber de antemão quais estão satisfeitos para entender e mitigar o que está acontecendo
 
Variáveis:
 - listadas...
 
Variáveis Respostas:
 - Exited:  Fecha conta nos próximos meses ou não? 

```python
df = pd.read_csv('Churn_Modelling.csv')
df
```

```python
df = df.drop('RowNumber', axis = 1) #joga fora a primeira coluna que não serve pra nada
df
```

```python
df = df.set_index('CustomerId') #coloca CustomerId como índice
df
```

```python
df.loc[15701354] #pega só os dados de um determinado cliente
```

```python
df.iloc[0] #pega os dados do primeiro índice
```

```python
df_d=pd.get_dummies(df)  #resolvendo o problema com textos
df_d
```

```python
df_d.iloc[:,0:20] #mostra só os primeiros índices ate 20
```

```python
x = df_d.drop('Exited', axis = 1) #todas as variáveis menos a coluna exited estão em x, que são as preditivas
y = df_d.Exited #variável resposta
```

```python
x
```

```python
y
```

```python
y.mean() #média de clientes que fecharam, 20%
```

```python
y.value_counts() #qtd de zeros e uns, 1 é quem fechou a conta
```

```python
y.value_counts()/y.shape[0] #outra forma de tirar a média
```

# Separação em Conjunto de Treino e de Teste

```python
from sklearn.model_selection import train_test_split

x_tr, x_ts, y_tr, y_ts = train_test_split(x,y, test_size = 0.30, random_state=61658)

#cria 4 variáveis de teste e treino e faz o teste em 30% dos dados aleatoriamente (70% treino)
```

```python
x_tr 
```

# Treinando Modelo

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=500, max_depth=10, random_state=61658, n_jobs=-1) 
rf
```

```python
rf.fit(x_tr, y_tr)
```

# Testando o Modelo

```python
rf.predict_proba(x_ts)
```

```python
pd.DataFrame(rf.predict_proba(x_ts), columns=["Propensão 0", "Propensão 1"])
```

```python
pred = rf.predict_proba(x_ts)[:, 1]
pred
```

# Avaliando o Modelo

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_ts, pred)
```

# Vendo Graficamente

```python
plt.figure(figsize=(15,5))
plt.hist(pred[y_ts==0], bins=30)

plt.figure(figsize=(15,5))
plt.hist(pred[y_ts==1], bins=30)
```

# Comparando Histogramas

```python
np.linspace(0,1,10) #explidita tamanho e quantidade dos bins
```

```python
plt.figure(figsize=(15,5))
plt.hist(pred[y_ts==0], bins=np.linspace(0, .5, 50), density=True)

plt.figure(figsize=(15,5))
plt.hist(pred[y_ts==1], bins=np.linspace(0, .5, 50), density=True)
```

Dá para ver que as pessoas com score mais altos(gráfico abaixo) são as que saem do banco

# Sobrepor Histogramas

```python
plt.figure(figsize=(15,5))
plt.hist(pred[y_ts==0], bins=np.linspace(0, .5, 50), density=True, color = 'g', alpha = .3)
plt.hist(pred[y_ts==1], bins=np.linspace(0, .5, 50), density=True, color = 'r', alpha = .3)
```

 Onde os histogramas se sobrepõem é onde o modelo não funciona bem
