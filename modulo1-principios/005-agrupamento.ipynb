```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#identificar padrão de compras de clientes
```

```python
df = pd.read_csv('base_cluster.csv')
df
```

```python
df2 = df.set_index('ID').filter(regex='Mnt', axis=1)
df2
```

```python
df2.sum(axis=1)
```

```python
#virando porcentual
df3=df2.divide(df2.sum(axis=1), axis='index')
df3
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=4)
km.fit(df3)
```

```python
clus = km.predict(df3)
clus
```

```python
np.unique(clus, return_counts=True)
```

```python
df3.loc[:, 'cluster'] = clus
df3
```

```python
df3.groupby('cluster').mean()
```

```python
aux = df3.groupby('cluster').mean()
aux
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,1,1, projection='polar')
plt.fill(aux.loc[0], alpha=.3)

plt.figure(figsize=(15,5))
plt.subplot(1,1,1, projection='polar')
plt.fill(aux.loc[1], color='r', alpha=.3)
```

```python
plt.figure(figsize=(15, 2*5))
color=['b', 'r', 'k', 'g']
for cl in aux.index:
    plt.subplot(2, 2, cl+1, projection='polar')
    plt.fill(aux.loc[cl], color=color[cl], alpha=.3)
    plt.title(cl, fontsize=18)
     
```
