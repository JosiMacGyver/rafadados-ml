```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('bcdata.sgs.27743.csv', sep=';')
df.loc [:,'data'] = pd.to_datetime(df.data, format= '%d/%m/%Y') #convertendo para o formato ano, mês e dia
df
```

```python
#plotar para ver série temporal para ver como se comporta

plt.figure(figsize=(15,5))
plt.plot(df.data, df.valor) #setar os valores do eixos x e y
```

```python

plt.figure(figsize=(15,5))
plt.plot(df.data, df.valor, ".-") #coloca um pontinho em cada ponto real e as linhas que os unem
plt.grid()
```

```python
#quero prever fevereiro de 2020 baseando em janeiro e tudo daqui pra baixo é da forma difícil
df.loc[df.data>"2020-01-01"] 
```

```python
df.loc[df.data>"2020-01-01"].valor.iloc[0] #dá o valor da primeira linha
```

```python
df.loc[df.data>"2020-01-01"].valor.loc[97] #dá o valor do índice 97
```

```python
predicoes = [] 
erros= []

for dat in df.data:
    if dat == pd.to_datetime("2012-01-01"):
        erros.append(np.nan)#não tem nada antes desse valor, então eu preciso colocar algo na lista
        predicoes.append(np.nan)
        continue
        
        
    valor = df.loc[df.data==dat].valor.iloc[0] #só o valor da primeira linha
    pred = df.loc[df.data==(dat-pd.DateOffset(months = 1))] .valor.iloc[0]#subtrai um mês (função pronta do pandas)
    
    predicoes.append(pred)
    erros.append(pred - valor)
    
    
```

```python
predicoes
```

```python
df_pred = pd.DataFrame() #funçao panda para criar tabela
df_pred.loc[:,'data'] = df.data 
df_pred.loc[:,'valor'] = df.valor
df_pred.loc[:,'predicoes'] = predicoes
df_pred.loc[:,'erros'] = erros

df_pred
```

```python
plt.figure(figsize=(15,5))
plt.plot(df_pred.data, df_pred.erros, ".-")
plt.grid()

#dá para ver nesse modelo que tirando um ponto em 2020 nada fica muito fora de zero
```

```python
plt.figure(figsize=(15,5))
plt.plot(df_pred.data, df_pred.erros, ".-")
plt.grid()
plt.xlim(left=pd.to_datetime('2020-01-01'))
```

```python
plt.figure(figsize=(15,5))
plt.plot(df_pred.data, df_pred.valor, ".-")#azul
plt.plot(df_pred.data, df_pred.predicoes, '.-')#amarelo
plt.grid()    

#a predição jogou a série um mês pra frente, porque acha que tudo é linear
```

```python
#agora da forma fácil, sem esse monte de for, condições etc. Agora para fazer o modelo Dummy
df.loc[:, 'predicoes'] = df.valor.shift(1) #cria coluna predições deslocando o valor de uma linha
df.loc[:, 'erros'] = df.predicoes - df.valor #calcula os erros
df
```

```python
plt.figure(figsize=(15,5))
plt.plot(df.data, df.valor, ".-")#azul
plt.plot(df.data, df.predicoes, '.-')#amarelo
plt.grid()    
```
