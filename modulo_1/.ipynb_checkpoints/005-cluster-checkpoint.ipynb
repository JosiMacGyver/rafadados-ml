```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Problema shopping

```python
df = pd.read_csv('base_cluster.csv')
df
```

```python
df2=df.set_index('ID').filter(regex='Mnt', axis=1) #filtra somente os dados que tem Mnt no nome
df2
```

```python
df2.sum(axis=1) #soma por linha para cada cliente
```

```python
df3=df2.divide(df2.sum(axis=1), axis='index') #divide por cada um dos itens da linha do cliente dando % de gasto total
df3
```

# Cluster

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=4)
km.fit(df3)
#faz 4 agrupamentos
```

```python
clus=km.predict(df3)
clus
#atrubiu um valor para cada um dos clientes: 0, 1, 2 ou 3
```

```python
np.unique(clus, return_counts=True)
```

```python
df3.loc[:, 'cluster'] = clus
df3
 
```

# Tirando informações da tabela

```python
df3.groupby('cluster').mean() #agrupa pelos índices iguais de cluster
```

```python
aux=df3.groupby('cluster').mean()
```

# Padrões e gráficos wtf!

```python
plt.figure(figsize=(15,5))
plt.subplot(1,1,1, projection='polar')
plt.fill(aux.loc[0], alpha=.5)

#visualização dos gastos do grupo 1
```

```python
plt.figure(figsize=(15,2*5)) #2 clusters em cima e dois em baixo
color = ['b', 'r', 'k','g']
for cl in aux.index:
    plt.subplot(2,2, cl+1, projection='polar')
    plt.fill(aux.loc[cl], color=color[cl], alpha=.5)
    plt.title(cl, fontsize=18)
    
```
