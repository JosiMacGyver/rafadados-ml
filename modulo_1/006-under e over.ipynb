```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
t = np.arange(0,7,.1) #array com tempo
t
```

```python
x = (0.9*9.8*t**2)/2.+10*np.random.randn(t.shape[0]) #array de posição com tempo segundo MUV
x
```

```python
plt.figure(figsize=(15,5))
plt.plot(t, x, 'o')
plt.xlabel('Tempo(s)')
plt.ylabel('Deslocamento(cm)')
```

# Regressão dos Dados

# Linear = Underfit

```python
#modelo mais simples de regressão: Linear
from sklearn.linear_model import LinearRegression 
lr = LinearRegression(normalize=True)
lr.fit(t.reshape(-1,1), x)
```

```python
t.reshape(-1,1); #reshape transforma o array em uma coluna
```

```python
pred=lr.predict(t.reshape(-1,1))
pred
```

```python
plt.figure(figsize=(15,5))
plt.plot(t, x, 'o')
plt.plot(t, pred, 'k.-')
plt.xlabel('Tempo(s)')
plt.ylabel('Deslocamento(cm)')
```

# Polinomial = Overfit

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree=200)
t_t = pf.fit_transform(t.reshape(-1,1))
t_t
```

```python
pd.DataFrame(t_t)
```

```python
lr = LinearRegression(normalize=True)
lr.fit(t_t, x)
```

```python
pred = lr.predict(t_t)
pred
```

```python
plt.figure(figsize=(15,5))
plt.plot(t, x, 'o')
plt.plot(t, pred, 'k-')
plt.xlabel('Tempo(s)')
plt.ylabel('Deslocamento(cm)')
```

# Forçando ser uma parábola, degree=2

```python
pf = PolynomialFeatures(degree=2, include_bias=False) #tira a primeira coluna de zeros
t_t = pf.fit_transform(t.reshape(-1,1))
```

```python
lr = LinearRegression(normalize=True)
lr.fit(t_t, x)
```

```python
pred = lr.predict(t_t)
```

```python
plt.figure(figsize=(15,5))
plt.plot(t, x, 'o')
plt.plot(t, pred, 'k-')
plt.xlabel('Tempo(s)')
plt.ylabel('Deslocamento(cm)')
```

```python
lr.coef_ #dá os coeficientes de t e t^2
```

Separar de novo em treino e teste

```python
from sklearn.model_selection import train_test_split
t_tr, t_ts, x_tr, x_ts = train_test_split(t, x, test_size=.3)
```

```python
plt.figure(figsize=(15,5))
plt.plot(t_tr, x_tr, 'bo') #variáveis de treino
plt.plot(t_ts, x_ts, 'ro') #variáveis de teste

#agora ele tem de aprender com as bolinhas azuis e testar se é válido nas vermelhas
```

```python
graus = range(1,30)
erros=[]

for n_pol in graus:
    pf = PolynomialFeatures(degree=n_pol)
    t_tr_t = pf.fit_transform(t_tr.reshape(-1,1))
    t_ts_t=pf.transform(t_ts.reshape(-1,1))
    
    lr=LinearRegression(normalize=True)
    lr.fit(t_tr_t, x_tr)
    
    pred = lr.predict(t_ts_t)
    
    #MAE (Mean Absolut Error)    
    erros.append(np.mean(np.abs(pred - x_ts))) 
```

Vendo a quatidade de erros em função de graus

```python
plt.figure(figsize=(15,5))
plt.plot(graus, erros)
plt.xticks(np.arange(30))
plt.grid()
```

```python
plt.figure(figsize=(15,5))
plt.plot(graus, erros)
plt.xlim(0,5)
plt.ylim(0,20)
```
