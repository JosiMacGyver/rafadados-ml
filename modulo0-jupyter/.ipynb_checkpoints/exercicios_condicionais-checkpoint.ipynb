## Avaliação de expressões <a id="expr" href="#topo">$\uparrow$</a>

> ### Instruções
> 
>  - Cada célula contém uma expressão condicional. Você deve avaliar se essa expressão retornará `True` ou `False`. 
>  - Faça isso em uma folha de papel antes de executá-la.
>  - As expressões estão ordenadas das mais simples para as mais complexas.
>  - Para as mais complexas, avalie cada expressão antes de avaliar os operadores lógicos `and` e `or`.

<hr />
<div style="float: left;">
    
### Operadores:
    
> | Operador | Significado    |
> |:--------:|:--------------:|
> |`>=`      | Maior ou igual |
> |`>`       | Maior          |
> |`==`      | Igual          |
> |`!=`      | Diferente      |
> |`<`       | Menor          |
> |`<=`      | Menor ou igual |
    
<br />
    
### Operadores lógicos
> #### and
> ``` python
> True  and True  == True
> True  and False == False
> False and True  == False
> False and False == False
> ```
>     
> #### or
> ``` python
> True  or True  == True
> True  or False == True
> False or True  == True
> False or False == False
> ```
</div>

````python
idade = 37

idade > 20
````

````python
idade = 42

idade >= 42
````

````python
idade = 28

idade > 28
````

````python
idade = 31

idade != 31
````

````python
idade = 42

(idade > 40) and (idade < 43)
````

````python
idade = 18

(idade > 15) or (idade < 20)
````

````python
idade = 18

(idade > 15) or (idade < 16)
````

````python
idade = 33

(idade < 30) and (idade > 32)
````

````python
idade = 40

(idade > 40) and (idade < 40)
````

````python
idade = 44

(idade < 20) or (idade == 44)
````

### Nota: Uma outra forma de escrever `and` e `or` <a href="#topo" id="outra_forma">$\uparrow$</a>

> - O operador `and` também pode ser escrito como `&`
> - O operador `or` também pode ser escrito como `|` (a tecla do lado do `Z` do seu teclado, pressionada com `Shift`)

````python
idade = 23

(idade > 20) & (idade < 30)
````

````python
idade = 49

(idade < 20) | (idade > 40)
````
